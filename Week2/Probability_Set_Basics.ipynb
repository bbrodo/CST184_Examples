```python

# --- Probability with Sets Example ---

# Total sample space: 100 countries
sample_space = set(range(1, 101))

def probability_summary(renewable_count, low_fossil_start, low_fossil_count=40):
    """Calculate probabilities for two country sets."""
    renewable = set(range(1, renewable_count + 1))
    low_fossil = set(range(low_fossil_start, low_fossil_start + low_fossil_count))
    both = renewable & low_fossil

    P_A = len(renewable) / len(sample_space)
    P_B = len(low_fossil) / len(sample_space)
    P_A_and_B = len(both) / len(sample_space)
    P_A_given_B = P_A_and_B / P_B

    return {
        "renewable countries": len(renewable),
        "low fossil countries": len(low_fossil),
        "overlap": len(both),
        "P(Renewable > 50%)": round(P_A, 2),
        "P(Fossil < 10%)": round(P_B, 2),
        "P(Both)": round(P_A_and_B, 2),
        "P(Renewable | Low Fossil)": round(P_A_given_B, 2),
    }

# Original example: 60 renewable countries and 40 low-fossil countries.
original = probability_summary(renewable_count=60, low_fossil_start=41)

print("Original example")
for label, value in original.items():
    print(f"{label}: {value}")

print("\nChanging the renewable range")
for renewable_count in [30, 60, 70]:
    result = probability_summary(renewable_count=renewable_count, low_fossil_start=41)
    print(
        f"Renewable={renewable_count:>2}, "
        f"overlap={result['overlap']:>2}, "
        f"P(A)={result['P(Renewable > 50%)']:.2f}, "
        f"P(A and B)={result['P(Both)']:.2f}, "
        f"P(A | B)={result['P(Renewable | Low Fossil)']:.2f}"
    )

print("\nChanging the overlap while keeping 60 renewable and 40 low-fossil countries")
for low_fossil_start in [61, 51, 41, 31, 21]:
    result = probability_summary(renewable_count=60, low_fossil_start=low_fossil_start)
    low_fossil_end = low_fossil_start + 39
    print(
        f"Low fossil={low_fossil_start:>2}-{low_fossil_end:<3}, "
        f"overlap={result['overlap']:>2}, "
        f"P(A and B)={result['P(Both)']:.2f}, "
        f"P(A | B)={result['P(Renewable | Low Fossil)']:.2f}"
    )

print("\nObservation")
print("More overlap increases P(Both) and P(Renewable | Low Fossil).")
print("Less overlap lowers both values, even when the individual set sizes stay the same.")
```
