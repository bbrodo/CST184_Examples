```python
import pandas as pd

df = pd.read_csv("energy_cleaned.csv")
df.head()
df.info()
df.describe()
```

Nothing unusual

```python
high_renew = df[df["%Renewable"] > 50]
print("Countries with >50% renewable energy:", len(high_renew))
```

30% of countries use mostly renewable energy

```python
region_avg = df.groupby("Region")["%Renewable"].mean().plot(kind='bar')
print(region_avg)
regions = df["Region"]
print(regions)
```

Brazil has the highest renewable energy use

```python
import matplotlib.pyplot as plt

df["%Renewable"].plot(kind="hist", bins=10, edgecolor="black")
plt.title("Histogram of Renewable Energy Use")
plt.xlabel("% Renewable")
plt.grid(True)
plt.show()
```

Most countries are clustered around 40%, 50%, and 90%

```python
df["%Renewable"].plot(kind="density", linewidth=2)
plt.title("Density Plot of Renewable %")
plt.xlabel("% Renewable")
plt.grid(True)
plt.show()
print(f"Mean: {df["%Renewable"].mean()}")
print(f"Median: {df["%Renewable"].median()}")
print(f"Mode(s): \n{df["%Renewable"].mode()}")
```

Distrubution is positively skewed and the frequency distrubution mode appears to be around the 45% mark, but if we are describing the frequency of each individual number then yes the data is multimodal

```python
test = df.groupby("Country")["%Renewable"].median()
test.plot(kind="bar", legend=True)
plt.title("Renewable Energy by Region")
plt.xlabel("% Renewable")
plt.grid(True)
plt.show()
```

The histogram tells us that globally, most countries do have around 50% renewable energy, but very few have >60%
South America and Asia are leading in renewable energy production
There is a surprisingly large gap between 50% and 90% renewable%
I think this analysis will help shape global energy policies by pointing out which countries are doing the best, and which need to improve.
