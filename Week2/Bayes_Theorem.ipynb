```python

# --- Bayes' Theorem Example in Python ---

# P(B | A): probability of heatwave on blackout days
P_heatwave_given_blackout = 0.80

def blackout_given_heatwave(P_blackout, P_heatwave):
    """Calculate P(blackout | heatwave) with Bayes' theorem."""
    return (P_heatwave_given_blackout * P_blackout) / P_heatwave

# Original values
P_blackout = 0.10  # P(A): probability of power blackout
P_heatwave = 0.30  # P(B): overall probability of heatwave

# Bayes' Theorem: P(A | B) = [P(B | A) * P(A)] / P(B)
P_blackout_given_heatwave = blackout_given_heatwave(P_blackout, P_heatwave)

print("Original values")
print("P(blackout):", P_blackout)
print("P(heatwave):", P_heatwave)
print("P(blackout | heatwave):", round(P_blackout_given_heatwave, 3))

print("\nChanging P_blackout while keeping P_heatwave = 0.30")
for test_blackout in [0.05, 0.10, 0.20]:
    result = blackout_given_heatwave(test_blackout, P_heatwave=0.30)
    print(f"P_blackout={test_blackout:.2f} -> P(blackout | heatwave)={result:.3f}")

print("\nChanging P_heatwave while keeping P_blackout = 0.10")
for test_heatwave in [0.20, 0.30, 0.50]:
    result = blackout_given_heatwave(P_blackout=0.10, P_heatwave=test_heatwave)
    print(f"P_heatwave={test_heatwave:.2f} -> P(blackout | heatwave)={result:.3f}")

print("\nObservation")
print("Increasing P_blackout raises the final probability.")
print("Increasing P_heatwave lowers the final probability because it is the denominator in Bayes' theorem.")
```

So the final result increases when P_blackout increases, because it is in the numerator. The final result decreases when P_heatwave increases, because it is in the denominator of Bayes’ theorem.
